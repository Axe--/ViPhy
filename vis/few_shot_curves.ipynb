# Few-Shot Curves

Plot model scores for `k` steps -- indicating % train set.

```python
import matplotlib.pyplot as plt
```

```python
bert = [55.56, 56.01, 55.96, 55.96, 56.08]  # zs = 31.90
rob  = [43.53, 53.53, 55.95, 55.95, 55.95]
deb  = [55.95, 55.95, 55.95, 55.95, 55.95]

# visb = [43.54, 55.95, 55.95, 55.95, 55.95]
vilt = [54.06, 49.81, 54.86, 54.90, 49.81]
flava= [41.98, 53.12, 56.69, 56.58, 56.61]
clip = [43.32, 55.33, 53.62, 56.79, 57.82]

cap  = [50.05, 49.12, 55.56, 55.58, 56.53]
```

```python

# Font
plt.rcParams.update({'font.size': 14})

# Style
l = 'solid'

# Line width
w = 2

# Figure
plt.figure(figsize=(12, 12))

# X-axis
x = [20, 40, 60, 80, 100]

# Plot
col_bt, = plt.plot(x, bert, label="BERT", linestyle=l, linewidth=w, color='orange')
col_rb, = plt.plot(x, rob,  label="RoBERTa", linestyle=l, linewidth=w, color='gold')
col_db, = plt.plot(x, deb,  label="DeBERTa", linestyle=l, linewidth=w, color='khaki')

col_vl, = plt.plot(x, vilt, label="ViLT", linestyle=l, linewidth=w, color='lightskyblue')
col_fl, = plt.plot(x, flava,label="FLAVA", linestyle=l, linewidth=w, color='turquoise')
col_cl, = plt.plot(x, clip, label="CLIP",   linestyle=l, linewidth=w, color='deepskyblue')

col_cb, = plt.plot(x, cap,  label="CapBERT", linestyle=l, linewidth=w, color='darkgray')

# Legend (models)
plt.legend()
# color_legend = plt.legend(handles=[col_cb, col_bt, col_cl], loc='lower right')
# plt.gca().add_artist(color_legend)

plt.xlabel('#Samples')
plt.xticks(x, fontsize=12)
plt.yticks(range(20, 61, 10), fontsize=12)
plt.grid(color='grey', linestyle=':', linewidth=0.5)

plt.show()
# plt.savefig('./few_shot.pdf', bbox_inches='tight')
```
